```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 30)

print('Libraries loaded ✅')
```

## 1. Load Raw Data

```python
df_raw = pd.read_csv('../data/raw/customers_dirty.csv')
print(f'Shape: {df_raw.shape}')
df_raw.head(10)
```

## 2. Data Quality Check

```python
# Missing values heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_raw.isnull(), cbar=True, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap (Yellow = Missing)', fontsize=14)
plt.tight_layout()
plt.savefig('../reports/missing_values_heatmap.png', dpi=150)
plt.show()
print('Missing values per column:')
print(df_raw.isnull().sum())
```

```python
# Check duplicates
print(f'Duplicate rows: {df_raw.duplicated().sum()}')
df_raw[df_raw.duplicated(keep=False)]
```

## 3. Run Cleaning Pipeline

```python
import sys
sys.path.append('../src')
from data_cleaning import *

# Run full pipeline
df = df_raw.copy()
df = handle_missing_values(df)
df = remove_duplicates(df)
df = fix_data_types(df)
df = handle_outliers(df)
df = standardize_text(df)
df = validate_email(df)
df = standardize_phone(df)

print('\nCleaning Complete!')
df.head()
```

## 4. After Cleaning Visualization

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Customer Data Analysis After Cleaning', fontsize=16)

# Age distribution
axes[0,0].hist(df['Age'], bins=10, color='steelblue', edgecolor='black')
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')

# Salary distribution
axes[0,1].hist(df['Salary'], bins=10, color='green', edgecolor='black')
axes[0,1].set_title('Salary Distribution')
axes[0,1].set_xlabel('Salary (BDT)')

# City count
city_counts = df['City'].value_counts()
axes[1,0].bar(city_counts.index, city_counts.values, color='coral')
axes[1,0].set_title('Customers by City')
axes[1,0].tick_params(axis='x', rotation=45)

# Gender distribution
gender_counts = df['Gender'].value_counts()
axes[1,1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%',
              colors=['#ff9999','#66b3ff'])
axes[1,1].set_title('Gender Distribution')

plt.tight_layout()
plt.savefig('../reports/data_analysis.png', dpi=150)
plt.show()
```

## 5. Save Cleaned Data

```python
df.to_csv('../data/cleaned/customers_cleaned.csv', index=False)
print('Cleaned data saved! ✅')
print(f'Final shape: {df.shape}')
df.info()
```
